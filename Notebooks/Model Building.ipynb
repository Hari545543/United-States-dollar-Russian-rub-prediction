```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import math
```

```python
X_train = pd.read_csv("/content/drive/MyDrive/Collab/USD | RUB prediction/X_train.csv")
X_train.head()
```

```python
X_test = pd.read_csv("/content/drive/MyDrive/Collab/USD | RUB prediction/X_test.csv")
X_test.head()
```

```python
Y_test = pd.read_csv("/content/drive/MyDrive/Collab/USD | RUB prediction/Y_test.csv")
Y_test.head()
```

```python
Y_train = pd.read_csv("/content/drive/MyDrive/Collab/USD | RUB prediction/Y_train.csv")
Y_train.head()
```

```python
def train_eval(model, x_train, x_test, y_train, y_test):
    
    model.fit(x_train,y_train)
    m = model.predict(x_test)
    print('R2 score: ',r2_score(y_test, m))
    print('Mean absolute error: ',mean_absolute_error(y_test, m))
    print('Mean square error: ',(mean_squared_error(Y_test, m))**(0.5))
    return model
```

```python
def plot_regression(x, y, prediction, i_name, d_name)->None:
    plt.scatter(x, y, color = 'green')
    plt.plot(x, prediction, color = 'red')
    plt.title(i_name + ' vs ' + d_name)
    plt.xlabel(i_name)
    plt.ylabel(d_name)
    plt.show()
```

```python
np.reshape(Y_train,(-1,1))
```

 **Support Vector Machine**

```python
SVR = train_eval(SVR(), X_train, X_test, Y_train, Y_test)
```

**Decision Tree**

```python
dtree = train_eval(DecisionTreeRegressor(), X_train, X_test, Y_train, Y_test)
```

**Saving the model**

By observing, R2 score,mean squared value, we can conclude that decision tree is much better than support vector machine. Hence, choosing Decision tree is much better.

```python
filename = '/content/drive/MyDrive/Collab/USD | RUB prediction/Model/USD|RUB_DT_model.pkl'
pickle.dump(dtree, open(filename, 'wb'))
```
